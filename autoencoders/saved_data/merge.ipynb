```python
import pandas as pd

# Read both files
df1 = pd.read_csv("matryoshka_f1_results.csv")
# df1 = pd.read_csv("matryoshka_results.csv")
df2 = pd.read_csv("06-10-chess-trained-results_with_01_2025_topk.csv")

# Convert columns to sets for comparison
cols1 = set(df1.columns)
cols2 = set(df2.columns)

# Find columns unique to each file
unique_to_file1 = cols1 - cols2
unique_to_file2 = cols2 - cols1
common_columns = cols1.intersection(cols2)

print(f"Total columns in file 1: {len(cols1)}")
print(f"Total columns in file 2: {len(cols2)}")
print(f"Number of common columns: {len(common_columns)}")
print("\nColumns unique to file 1:")
for col in sorted(unique_to_file1):
    print(f"- {col}")
print("\nColumns unique to file 2:")
for col in sorted(unique_to_file2):
    print(f"- {col}")

# Optionally, if you want to see sample values from mismatched columns
if len(unique_to_file1) > 0:
    print("\nSample values from columns unique to file 1:")
    for col in sorted(unique_to_file1):
        sample_vals = df1[col].head(3)
        print(f"\n{col}:")
        print(sample_vals)
```

```python
import pandas as pd

# Read both files
df1 = pd.read_csv("matryoshka_f1_results.csv")
df2 = pd.read_csv("06-10-chess-trained-results_with_01_2025_topk.csv")

# Reindex df1 to have all columns from df2, filling missing ones with 0
df1_expanded = df1.reindex(columns=df2.columns, fill_value=0)

# Concatenate the dataframes
merged_df = pd.concat([df1_expanded, df2], axis=0, ignore_index=True)

# Save the merged result
merged_df.to_csv("merged_results.csv", index=False)

print(f"\nOriginal shapes:")
print(f"File 1: {df1.shape}")
print(f"File 2: {df2.shape}")
print(f"Merged file: {merged_df.shape}")

# Optional: verify no columns were lost
print(f"\nFinal number of columns: {len(merged_df.columns)}")
print(f"Should match file 2's columns: {len(merged_df.columns) == len(df2.columns)}")
```

```python
import pandas as pd

# Read the merged file
merged_df = pd.read_csv("merged_results.csv")

# Print shape before dropping
print(f"Original shape: {merged_df.shape}")

# Drop rows where trainer_class is 'GatedAnnealTrainer'
filtered_df = merged_df[merged_df['trainer_class'] != 'GatedAnnealTrainer']

# Print shape after dropping
print(f"Shape after dropping GatedAnnealTrainer: {filtered_df.shape}")

# Verify the trainer classes that remain
print("\nRemaining trainer classes:")
print(filtered_df['trainer_class'].unique())

# Save the filtered dataset
filtered_df.to_csv("merged_results.csv", index=False)
```
